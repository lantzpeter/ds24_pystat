### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

Objektorienterad approach eller pyplot/MATLAB style.

Den rekommenderade är objektorienterad. Speciellt när man skall göra mer avancerade och större projekt. För mindre små, snabba "plots" kan man använda den andra metoden.

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

Figur: Figuren hanterar alla grafer, titlar, förklaringar, färgskalor och andra detaljer, samt eventuella mindre figurer inuti den.

Axes: Det är hela området där grafen ritas, och inkluderar x- och y-axlar, själva plottningen, titlar, etiketter och tickmarkeringar.

Axis: Det är de enskilda axlarna inom ett Axes-objekt som ansvarar för skalor, tickmarkeringar och etiketter längs en viss riktning (x, y eller z).

Artist: Det är i princip är allt du ser på en figur, inklusive själva figuren, samt Axes- och Axis-objekten.

### When plotting in Matplotlib, what is the expected input data type?

Den indata som förväntas är i form av numpy.array, numpy.ma.masked_array eller objekt som kan omvandlas till en numpy.asarray.

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
x = np.linspace(-4, 4, 100)
y = x**2
```

```python
x
```

```python
y
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)

ax.set(title="Values of x^2 between -4 to 4", xlabel="x", ylabel="y")
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 4))
axs[0].scatter(x, y)

axs[1].bar(names, values)

axs[0].set(title="Values of x^2 between -2 to 2", xlabel="x", ylabel="y")
axs[1].set(title="Just some random fruits!", xlabel="Kind of fruit", ylabel="Quantity")
fig.suptitle("One Figure with 2 subplots")
```
